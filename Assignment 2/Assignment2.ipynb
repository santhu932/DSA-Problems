```python
def postfix_string(arr):
    if len(arr) == 0:
        return None
    st = []
    for i in range(len(arr)):
        if str(arr[i]) != '+' and str(arr[i]) != '-' and str(arr[i]) != '*':
            st.append(arr[i])
        if arr[i] == '+':
            if len(st) > 1 :
                s = st.pop()
                f = st.pop()
                st.append(f + s)
        if arr[i] == '-':
            if len(st) > 1:
                s = st.pop()
                f = st.pop()
                st.append(f[:int(s)] + f[int(s)+1:])
        if arr[i] == '*':
            if len(st) > 1:
                s = st.pop()
                f = st.pop()
                st.append(f * int(s))
    result = st.pop()        
    return result
```

```python
class ListNode():
      def __init__(self, val, next=None):
            self.val = val
            self.next = next

class AstronomicalInt():
    
    def insert(self, head, element):
        node = ListNode(element)
        if head == None:
            head = node
        else:
            p = head
            while p.next != None:
                p = p.next
            p.next = node
        return head
    
    def reverse(self, head):
        if head == None or head.next == None:
            return head
        present = head
        back = None
        front = None
        while present != None:
            front = present.next
            present.next = back
            back = present
            present = front
        head = back
        return head
    
    def convert(self, num:str):
        if num == '':
            return None
        head = None
        for k in range(len(num)):
            head = self.insert(head, num[k])
        return head
            
    def add(self, num1:ListNode, num2: ListNode):
        if num1 == None:
            return num2
        if num2 == None:
            return num1
        num1 = self.reverse(num1)
        num2 = self.reverse(num2)
        result = None
        last = None
        s = 0
        c = 0
        
        while num1 != None or num2 != None:
            if num1 != None:
                data1 = int(num1.val)
            else:
                data1 = 0
                
            if num2 != None:
                data2 = int(num2.val)
            else:
                data2 = 0
                
            s = data1 + data2 + c
            if s >= 10:
                c = 1
            else:
                c = 0
            
            s = s % 10
            node = ListNode(s)
            if result == None:
                result = node
            else:
                last.next = node
            last = node
            
            if num1 != None:
                num1 = num1.next
            if num2 != None:
                num2 = num2.next
            
        if c > 0:
            last.next = ListNode(c)

        result = self.reverse(result)
        return result

    def to_string(self, head):
        num = []
        curr = head
        while curr:
            num.append(curr.val)
            curr = curr.next
        num = ''.join([str(i) for i in num])
        return num
```

```python
def combination_check(n,st):
    number = str(n)
    words = st.split()
    match_dict = dict()
    w = []
    if len(number) != len(words):
        return False
    for m in range(len(words)):
        if number[m] not in match_dict:
            if words[m] not in w:
                match_dict[number[m]] = words[m]
                w.append(words[m])
            else:
                return False
        else:
            if words[m] != match_dict[number[m]]:
                return False
    return True
```

```python
import collections
def countAtoms(formula):
    size = len(formula)
    stk = [collections.Counter()]
    c = 0
    while c < size:
        if formula[c] == '(':
            stk.append(collections.Counter())
            c = c + 1
        elif formula[c] == ')':
            element = stk.pop()
            c = c+1
            k = c
            while c < size and formula[c].isdigit():
                c = c + 1
            if formula[k:c] != '':
                factor = int(formula[k:c])
            else:
                factor = 1
            for k, v in element.items():
                stk[-1][k] = stk[-1][k] + (v*factor)
        else:
            k = c
            c = c+ 1
            while c < size and formula[c].islower():
                c = c+1
            element1 = formula[k:c]
            k = c
            while c < size and formula[c].isdigit():
                c = c + 1
            if formula[k:c] != '':
                factor = int(formula[k:c])
            else:
                factor = 1
            stk[-1][element1] = stk[-1][element1] + factor
    string = ""
    for element in sorted(stk[-1]):
        if stk[-1][element] > 1:
            string = string + element + str(stk[-1][element])
        else:
            string = string + element + ''
            
    return string
```
